```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
dig =load_digits()
dir(dig)
```

```python
print(dig.data[0],dig.target[0],dig.target_names[0])
```

```python
# plt.gray()
plt.matshow(dig.images[1])
```

```python
X_train,X_test, y_train, y_test = train_test_split(dig.data,dig.target,test_size =0.1)
```

```python
mdl = LogisticRegression()
```

```python
mdl.fit(X_train,y_train)
```

```python
mdl.score(X_test,y_test)
```

```python
dig.images[12]
```

```python
plt.matshow(dig.images[12])
print("It is :",dig.target[12])
```

```python
mdl.predict([dig.data[12]])
```

```python
y_pred = mdl.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
